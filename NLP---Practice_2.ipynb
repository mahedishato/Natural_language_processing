```python
import nltk 
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

## tokenization paragraph-->sentences--words

```python
nltk.download('punkt')

sentence=nltk.sent_tokenize(paragrap)
```

```python
print(sentence)
```

```python
type(sentence)
```

```python
stemmer = PorterStemmer()
```

```python
stemmer.stem()
```

## now we can also use lemetization

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemetizer = WordNetLemmatizer()
```

```python
len(sentence)
```

## cleaning the datas

```python
import re
```

```python
corpus = []
for i in range(len(sentence)):
    text=re.sub('[^a-zA-Z]', ' ', sentence[i])
    text= text.lower()
    corpus.append(text)
    
```

```python
type(corpus)
```

```python
corpus
```

```python
import nltk
nltk.download('all')
```

## applying stemming 

```python
for i in corpus:
    words= nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(stemmer.stem(word))
```

## lemetization

```python
for i in corpus:
    words= nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(lemetizer.lemmatize(word))
```
